```python
import random
import time
import numpy as np
import matplotlib.pyplot as plt

# Funcin de Insertion Sort (O(n^2) en el peor caso, O(n) en el mejor caso)
def insertion_sort(arr):
    for j in range(1, len(arr)):
        key = arr[j]
        i = j - 1
        while i >= 0 and arr[i] > key:
            arr[i + 1] = arr[i]
            i -= 1
        arr[i + 1] = key
    return arr

# Generar datos de prueba

# Mejor caso-Funcin de InsertSort con lista en orden (mejor caso O(n))
def best_case(n):
    return list(range(n))  # Lista ordenada

#Caso Promedio - Funcin de InsertSort con lista aleatoria (caso O(n^2))
def average_case(n):
    return random.sample(range(n), n)  # Lista aleatoria

#Peor Caso - Funcin de InsertSort con lista inversa (peor caso O(n^2))
def worst_case(n):
    return list(range(n, 0, -1))  # Lista en inversa

# Medir tiempos de ejecucin
def measure_execution_time(func, arr):
    start_time = time.time()
    func(arr.copy())  # Copia para evitar modificar el original
    return time.time() - start_time

# Graficar resultados
def plot_results():
    sizes = list(range(100, 10001, 1000))  # De 100 a 10,000 con saltos de 1000
    times_best = []
    times_avg = []
    times_worst = []

    for size in sizes:
        times_best.append(measure_execution_time(insertion_sort, best_case(size)))
        times_avg.append(measure_execution_time(insertion_sort, average_case(size)))
        times_worst.append(measure_execution_time(insertion_sort, worst_case(size)))

    plt.plot(sizes, times_best, label='Best Case (O(n))', color='green')
    plt.plot(sizes, times_avg, label='Average Case (O(n))', color='blue')
    plt.plot(sizes, times_worst, label='Worst Case (O(n))', color='red')
    plt.xlabel('Size of Input')
    plt.ylabel('Time (seconds)')
    plt.legend()
    plt.title('Insertion Sort Time Complexity')
    plt.grid(True)
    plt.show()

# Ejecutar y graficar
plot_results()
```
